### Usage example

Test image:



First, need to instantiate the image classifier.

```python
from image_classification import ImageClassifier

ic = ImageClassifier(model_path="EfficientNetV2S_100epochs_0p8dropout.keras",
                     class_names=['Battery', 'Camera', 'Keyboard', 'Laptop', 'Microwave', 'Mobile', 'Mouse', 'PCB', 'Player', 'Printer', 'Smartwatch', 'Television', 'Washing Machine'])
```

Then proceed. Example usage:

```python
pred, confidence = ic.predict_class('galaxy_z_fold_7.webp')
print(pred)
```

NB: first prediction will be slow, subsequent predictions much quicker
